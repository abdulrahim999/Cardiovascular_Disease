```python
# lodaing my data which is cardio

df = pd.read_csv("cardio.csv" , sep=";")

df
```

```python
# we remove ID colmen

df = df.drop('id', axis=1)
df.head()
```

```python
# we have the age callculated in dayes and flot
# we want to convert it in years with round up to be int

df.age.describe()
```

```python
df["age"] = round(df["age"] / 365)
df
```

```python
df.describe()
```

```python
# to minimies the zeroes after the comma .0000000
pd.set_option('display.float_format',lambda x:'%.2f'% x)
```

```python
df.describe()
```

```python
# we found outlyer in ap_hi with max 16020 and min -150  so we have to fix this issue with to be in the range of 108 and 200

df['ap_hi']  = df['ap_hi'].apply(lambda x: (np.nan if x <= 108 else x))
df['ap_hi']  = df['ap_hi'].apply(lambda x: (np.nan if x >= 200 else x))
```

```python
df.describe()
```

```python
# from the describe we found that ap_lo  have outlyer with max 11000 and min -70 we clean this with the range of 76 and 130

df['ap_lo']  = df['ap_lo'].apply(lambda x: (np.nan if x <= 76 else x))
df['ap_lo']  = df['ap_lo'].apply(lambda x: (np.nan if x >= 130 else x))
```

```python
df.describe()
```

```python
# from the describe we found that ap_lo  have outlyer with max 250 and min 8,21 we clean this with the range of 150 and 190

df['height']  = df['height'].apply(lambda x: (np.nan if x <= 150 else x))
df['height']  = df['height'].apply(lambda x: (np.nan if x >= 190 else x))
```

```python
df.describe()
```

```python
# from the describe we found that ap_lo  have outlyer with max 200 and min 11 we clean this with the range of 50 and 140

df['weight']  = df['weight'].apply(lambda x: (np.nan if x <= 50 else x))
df['weight']  = df['weight'].apply(lambda x: (np.nan if x >= 140 else x))
```

```python
df.describe()
```

```python
# we have the gender female reprsntitive 1 and male by 2

# we can convert it to male and female insted of 2,1

# we will use this outcome in our statices 

df.gender.value_counts()
```
